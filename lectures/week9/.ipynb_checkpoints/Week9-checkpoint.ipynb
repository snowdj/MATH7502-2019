Look up ? svd later...

```julia
F = svd(A)
```

```julia
U = F.U;
V = F.V; #can also use Vt
Σ = Diagonal(F.S)
maximum(abs.(A - U*Σ*V'))
```

```julia
scatter(F.S[2:50],legend=false)
```

```julia
S = F.S;
```

```julia
approx(k) = sum([S[i]*U[:,i]*V[:,i]' for i in 1:k])
```

```julia
B1 = approx(2);
heatmap(B1,yflip=true)
```

# Least Squares

```julia
A = [ 2 0 ; - 1 1; 0 2];
b = [1,0,-1];
xhat = A\b
```

```julia
norm(A*xhat-b)
```

```julia
loss(x1,x2) = norm(A*[x1,x2]-b)
```

```julia
loss(xhat...)
```

```julia
loss(xhat + 0.01*rand(2)...)
```

```julia
pp =inv(A'*A)*A';
pp*b
```
